```python
# Import the modules we created
from parse_fasta import parse_fasta_file
from esm_api import predict_sequences

# Specify your FASTA file path here
fasta_file_path = "your_sequences.fasta"  # Change this to your FASTA file path

# Parse FASTA file
try:
    sequences = parse_fasta_file(fasta_file_path)
    print(f"Loaded {len(sequences)} sequences from {fasta_file_path}")
    for i, seq in enumerate(sequences, 1):
        print(f"  Sequence {i}: {len(seq)} amino acids")
except FileNotFoundError:
    print(f"FASTA file not found: {fasta_file_path}")
    print("Using example sequences instead...")
    # Fallback to example sequences if file not found
    sequences = ["ACDG<mask>HIKLMN", "XPQRS<mask>FGT"]

# Make predictions using ESM API
print(f"\nSending {len(sequences)} sequences to API...")
try:
    results = predict_sequences(sequences)
    print("\nPredictions received:")
    print(results)
except Exception as e:
    print(f"Error: {e}")
```
